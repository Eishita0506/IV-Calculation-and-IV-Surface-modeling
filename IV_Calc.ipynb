# Calculation of Implied Volatility from Option Prices

```python
import numpy as np
from scipy.stats import norm
```

1. C=Market Price
2. S= Spot Price
3. T= Time to Maturity
4. r= risk free interest rate
5. sigma= Volatility

# Call Option

```python
def black_scholes_call(S,K,T,r, sigma):
    d1=(np.log(S/K)+(r+0.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d2=d1-sigma*np.sqrt(T)
    call_option=S*norm.cdf(d1)-K*np.exp(-r*T)*norm.cdf(d2)
    return call_option
```

Newton Raphson Method

```python
def vega(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + sigma ** 2 / 2) * T) / (sigma * np.sqrt(T))
    vega = S  * np.sqrt(T) * norm.pdf(d1)
    return vega
```

```python
def newton_IV_call(C, S, K, T, r, tol=0.000001,
                            max_iterations=200):
    sigma = 0.2
    for i in range(max_iterations):
        eqn = black_scholes_call(S, K, T, r, sigma) - C
        if abs(eqn) < tol:
            break
        sigma = sigma - eqn / vega(S, K, T, r, sigma)

    return sigma
```

```python
C= 6.5
S = 100
K = 105
T = 0.5
r = 0.05

imp_volatility = newton_IV_call(C, S, K, T, r)
print('Implied volatility of the call option using Newton Rapshon is: ',imp_volatility)
```

Bisection Method

```python
def bisection_IV_call(C, S, K, T, r, low_bound, up_bound,  tolerance=0.000001, max_iterations=200):
    for i in range(max_iterations):
        mid=(low_bound+up_bound)/2
        price_mid=black_scholes_call(S,K,T,r,mid)
        diff= price_mid-C
        if abs(diff)< tolerance:
            return mid
        
        price_low=black_scholes_call(S,K,T,r,low_bound)
        diff_low=price_low-C
        if diff_low*diff<0:
            up_bound=mid
        else:
            low_bound=mid
    return (up_bound + low_bound)/2          
```

```python
C= 6.5
S = 100
K = 105
T = 0.5
r = 0.05

imp_volatility = bisection_IV_call(C, S, K, T, r, 0.000001, 5)
print('Implied volatility of the call option using bisection is: ',imp_volatility)
```

# Put Option

```python
def black_scholes_put(S,K,T,r, sigma):
    d1=(np.log(S/K)+(r+0.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d2=d1-sigma*np.sqrt(T)
    put_option=K*np.exp(-r*T)*norm.cdf(-d2)-S*norm.cdf(-d1)
    return put_option
```

Newton Raphson Method

```python
def vega(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + sigma ** 2 / 2) * T) / (sigma * np.sqrt(T))
    vega = S  * np.sqrt(T) * norm.pdf(d1)
    return vega
```

```python
def newton_IV_put(C, S, K, T, r, tol=0.000001,
                            max_iterations=200):   
    sigma = 0.2
    for i in range(max_iterations):
        eqn = black_scholes_put(S, K, T, r, sigma) - C
        if abs(eqn) < tol:
            break
        sigma = sigma - eqn / vega(S, K, T, r, sigma)

    return sigma
```

```python
C= 9.2
S = 100
K = 105
T = 0.5
r = 0.05

imp_volatility = newton_IV_put(C, S, K, T, r)
print('Implied volatility of the put option using Newton Rapshon is: ',imp_volatility)
```

Bisection Method

```python
def bisection_IV_put(C, S, K, T, r, low_bound, up_bound,  tolerance=0.000001, max_iterations=200):
    for i in range(max_iterations):
        mid=(low_bound+up_bound)/2
        price_mid=black_scholes_put(S,K,T,r,mid)
        diff= price_mid-C
        if abs(diff)< tolerance:
            return mid
        
        price_low=black_scholes_put(S,K,T,r,low_bound)
        diff_low=price_low-C
        if diff_low*diff<0:
            up_bound=mid
        else:
            low_bound=mid
    return (up_bound + low_bound)/2   
```

```python
C= 9.2
S = 100
K = 105
T = 0.5
r = 0.05

imp_volatility = newton_IV_put(C, S, K, T, r)
print('Implied volatility of the put option using Bisection is: ',imp_volatility)
```
