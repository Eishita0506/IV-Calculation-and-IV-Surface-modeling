```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker=yf.Ticker("^SPX")
expiry_dates=ticker.options
print(expiry_dates)
chain=ticker.option_chain(expiry_dates[0])
calls=chain.calls
puts=chain.puts
```

```python
print("Calls:")
calls.head()
```

```python
print("Puts:")
puts.head()
```

```python
otm_calls=calls[calls['inTheMoney']==False]
otm_calls.head()
```

```python
otm_puts=puts[puts['inTheMoney']==False]
otm_puts.head()
```

```python
plt.figure(figsize=(14,9))
plt.plot(otm_calls['strike'], otm_calls['impliedVolatility'], label='OTM Calls', marker='o')
plt.plot(otm_puts['strike'], otm_puts['impliedVolatility'], label='OTM Puts', marker='x')
plt.xlabel("Strike Price($)")
plt.ylabel("Implied Volatility")
plt.title("Implied Volatility Vs Strike Price for SPX")
plt.grid(True, alpha=0.2)
plt.legend()
plt.show
```

```python
all_options=[]
expiries=ticker.options
print("Total Expiries: ", len(expiries))
```

```python
for expiry in expiries:
    chain=ticker.option_chain(expiry)
    calls=chain.calls.copy()
    puts=chain.puts.copy()
    calls['expiry']=expiry
    puts['expiry']=expiry
    calls['type']="Call"
    puts['type']="Put"
    all_options.append(calls)
    all_options.append(puts)
all_options=pd.concat(all_options, ignore_index=True) 
all_options.head()   
```

```python
index_to_drop=all_options[all_options['inTheMoney']==True].index
all_options=all_options.drop(index_to_drop)
all_options.head()
```

```python
expirations=all_options['expiry'].unique()
plt.figure(figsize=(12,6))
for exp in expirations:
    df_exp=all_options[all_options['expiry']==exp]
    all_otm_calls=df_exp[df_exp['type']=='Call']
    all_otm_puts=df_exp[df_exp['type']=='Put']
    plt.plot(all_otm_calls['strike'], all_otm_calls['impliedVolatility'], marker='o')
    plt.plot(all_otm_puts['strike'], all_otm_puts['impliedVolatility'], marker='x')
plt.title("Volatility Smile for All Expiration")
plt.xlabel('Strike Price($)')
plt.ylabel('Implied Volatility')
plt.legend()
plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
expirations=all_options['expiry'].unique()
for exp in expirations:
    df_exp=all_options[all_options['expiry']==exp]
    plt.figure(figsize=(12,6))
    all_otm_calls=df_exp[df_exp['type']=='Call']
    all_otm_puts=df_exp[df_exp['type']=='Put']
    plt.plot(all_otm_calls['strike'], all_otm_calls['impliedVolatility'], marker='o', label='OTM Calls')
    plt.plot(all_otm_puts['strike'], all_otm_puts['impliedVolatility'], marker='x', label='OTM Puts')
plt.title(f"Volatility Smile for Expiration {exp}")
plt.xlabel('Strike Price($)')
plt.ylabel('Implied Volatility')
plt.legend()
plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
today=pd.Timestamp.today()
df=[]
expiries=ticker.options
for expiry in expiries:
    chain=ticker.option_chain(expiry)
    calls=chain.calls.copy()
    puts=chain.puts.copy()
    calls['expiry']=expiry
    puts['expiry']=expiry
    calls['type']="Call"
    puts['type']="Put"
    df.append(calls)
    df.append(puts)
df=pd.concat(df, ignore_index=True) 
df['expiry']=pd.to_datetime(df['expiry'])
df['Maturity']=(df['expiry']-today).dt.days/252
df_calls=df[df['type']=='Call']
df_puts=df[df['type']=='Put']
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(df_calls['strike'], df_calls['Maturity'], df_calls['impliedVolatility'], cmap='viridis', edgecolor='none', alpha=0.7, label='Calls')
ax.plot_trisurf(df_puts['strike'], df_puts['Maturity'], df_puts['impliedVolatility'], cmap='plasma', edgecolor='none', alpha=0.7, label='Puts')
ax.set_xlabel('Strike Price')
ax.set_ylabel('Time to Expiry (Years)')
ax.set_zlabel('Implied Volatility')
ax.set_title('3D Implied Volatility Surface for SFX Options')
handles = [Line2D([], [], marker='o', linestyle='None', color='blue', label='Calls'),
           Line2D([], [], marker='o', linestyle='None', color='red', label='Puts')]
ax.legend(handles=handles)
plt.show()
```
